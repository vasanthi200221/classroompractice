# PCA Assignment

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
wine=pd.read_csv("wine.csv")
wine
```

```python
wine1=wine.iloc[:,1:]
wine1
```

```python
wine1.describe()
```

```python
wine1.info()
```

```python
wine["Type"].value_counts()
```

```python
wine.shape
```

```python
# Converting data to numpy array
```

```python
wine_arr=wine1.values
```

```python
wine_arr
```

```python
# Normalizing the  numerical data
wine_norm=scale(wine_arr)
wine_norm
```

```python
# Applying PCA Fit Transform to dataset
```

```python
pca=PCA(n_components=13)
pca_values=pca.fit_transform(wine_norm)
pca_values
```

```python
# PCA Components matrix or convariance Matrix
pca.components_
```

```python
# The amount of variance that each PCA has
var = pca.explained_variance_ratio_
var
```

```python
# Cummulative variance of each PCA
Var = np.cumsum(np.round(var,decimals= 4)*100)
Var
```

```python
plt.plot(Var,color="blue");
```

```python
# Final Dataframe
final_df=pd.concat([wine['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
import seaborn as sns
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=final_df)
```

```python
sns.scatterplot(data=final_df, x='PC1', y='PC2', hue='Type');
```

#  Clustering with original data

```python
# Heirarchical clustering
```

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch # for creating dendrogram
#p = np.array(df_norm) # converting into numpy array format
z = linkage(wine_norm , method="average",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=6.,  # rotates the x axis labels
    #leaf_font_size=15.,  # font size for the x axis labels
)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import warnings
warnings.filterwarnings('ignore')
h_complete = AgglomerativeClustering(n_clusters=5, linkage='average',affinity = "euclidean").fit(wine_norm )

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
wine['clust']=cluster_labels # creating a  new column and assigning it to new column
wine
```

```python
wine.iloc[:,1:].groupby(wine.clust).mean()
```

```python
# Kmeans clustering
```

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(wine_norm)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()
```

```python
clf = KMeans(n_clusters=3)
y_kmeans = clf.fit_predict(wine_norm)
```

```python
y_kmeans
```

```python
clf.cluster_centers_
```

```python
clf.inertia_
```

```python
md=pd.Series(y_kmeans)  # converting numpy array into pandas series object
wine['clustk']=md # creating a  new column and assigning it to new column
wine
```
