```python
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
book=pd.read_csv("book.csv")
```

```python
book
```

# Apriori Algorithm

 Association rules with 10% Support and 70% confidence

```python
frequent_itemsets=apriori(book,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
rules
```

```python
rules.sort_values('lift',ascending=False)
```

An leverage value of 0 indicates independence. Range will be [-1 1]
A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```

```python
# visualization of obtained rule
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

```python
# visualization of obtained rule
plt.scatter(rules['support'],rules['lift'])
plt.xlabel('support')
plt.ylabel('lift')
plt.show()
```

# Association rules for My Movies data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
movie=pd.read_csv("my_movies.csv")
movie
```

```python
movie.shape
```

```python
movie.info()
```

```python
movie2=movie.iloc[:,5:]
```

```python
movie2
```

# Apriori Algorithm

1. Association rules with 10% Support and 70% confidence

```python
frequent_itemsets=apriori(movie2,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
rules
```

```python
rules[rules.lift>1]
```

```python
# visualization of obtained rule
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

```python
# with 5% support
frequent_itemsets2=apriori(movie2,min_support=0.05,use_colnames=True)
frequent_itemsets2
```

```python
# 90% confidence
rules2=association_rules(frequent_itemsets2,metric='lift',min_threshold=0.9)
rules2
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
rules2[rules2.lift>1]
```

```python
# visualization of obtained rule
plt.scatter(rules2['support'],rules2['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
