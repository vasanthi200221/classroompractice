# SVM forest fires Assignment

```python
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.svm import SVC
from sklearn.model_selection import train_test_split,cross_val_score

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
forestfires = pd.read_csv("forestfires.csv")
```

```python
forestfires
```

```python
data = forestfires.describe()
data
```

```python
##Dropping the month and day columns
forestfires.drop(["month","day"],axis=1,inplace =True)
```

```python
forestfires
```

```python
predictors = forestfires.iloc[:,0:28]
target = forestfires.iloc[:,28]
```

```python
predictors
```

```python
target
```

```python
##Normalising the data as there is scale difference
def norm_func(i):
    x= (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
fires = norm_func(predictors)
```

```python
fires
```

```python
x_train,x_test,y_train,y_test = train_test_split(predictors,target,test_size = 0.25, stratify = target)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf = SVC()
param_grid = [{'kernel':['rbf','linear','sigmoid','poly'],'gamma':[50,5,10,0.5],'C':[15,14,13,12,11,10,0.1,0.001] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_
```

```python
model_linear = SVC(kernel = "linear",C=0.1,gamma=50,random_state=42)
model_linear.fit(x_train,y_train)
```

```python
pred_test_linear = model_linear.predict(x_test)
```

```python
np.mean(pred_test_linear==y_test)
```

```python
acc = accuracy_score(y_test, pred_test_linear) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, pred_test_linear)
```

```python
# Kernel = poly
model_poly = SVC(kernel = "poly",C=0.1,gamma=50,random_state=42)
model_poly.fit(x_train,y_train)
```

```python
pred_test_poly = model_poly.predict(x_test)
```

```python
np.mean(pred_test_poly==y_test)
```

```python
# kernel = rbf
model_rbf = SVC(kernel = "rbf",C=0.1,gamma=50,random_state=42)
model_rbf.fit(x_train,y_train)
```

```python
pred_test_rbf = model_rbf.predict(x_test)
```

```python
np.mean(pred_test_rbf==y_test)
```

```python
#'sigmoid'
model_sig = SVC(kernel = "sigmoid",C=0.1,gamma=50,random_state=42)
model_sig.fit(x_train,y_train)
```
